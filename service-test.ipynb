Imos comprobar se se fixo apresto dos servizos dos contedores.
Para elo emprega:
- Os métodos ```socket``` dentro dunha estrutura tipo```try: except:```.
- Un sinal ((*flag*) para indicaren se tódolos servizos están aprestos ou dispoñíbeis.
- Unha función de comprobación que emprega un dicionario cos servizos a comprobar, de xeito que o código é actualizábel e mais reutilizábel.
- No caso dunha excepción móstrase o erro para axudar na resulución de problemas (*troubleshooting*): <br>Se ben a saída soe ser ```Connection refused```.<br>--> Recoméndase ir á terminal e executar ```sudo docker ps``` para sabermos se o contedor chegou a cargar. Tamén se aconsella revisa-lo arquivo de configuración ```docker-compose.yml``` .

````python

# Libraría para facer conexións tipo SOCKET

import socket

# Libraría de funcións de sistema. Neste caso para empregarmos a estrutura "try-except-finally"

import sys

# Función que comproba se o porto de servizo está aberto ou que leria.

def comprobar_porto(host, porto, nome_servizo):
    # Tenta abrir un socket para comprobar se os portos de servizo están abertos.
    sock = socket.socket(socket.AF_INET, socket.SOCK_STREAM)
    sock.settimeout(1) # Tempo de espera
    
    try:
        # Tenta conectar ao porto
        sock.connect((host, porto)) 
        print(f"✅ {nome_servizo} ({porto}): DISPOÑÍBEL")
        return True
    except socket.error as e: # Non hai dios
        print(f"❌ {nome_servizo} ({porto}): NON DISPOÑÍBEL ({e})")
        return False
    finally:  # Hai que pechar o socket 
        sock.close()

# Dicionario de servizos e portos a comprobar

SERVIZOS = {
    "MySQL": 3306,
    "Cassandra": 9042,
    "Redis": 6379,
    "MongoDB": 27017
}

host_local = "127.0.0.1" # Direción IP de seu (localhost)

# Executáremolos a comprobación
print("Estado dos Servizos de Docker\n==============================")

# Facer apresto o sinal de éxito rotundo (success flag initialisation)
todos_dispoñibles = True

# Iteramos comprobar_porto. Se algún falla actualizamos o sinal de éxito rotundo, que xa non é.
for nome, porto in SERVIZOS.items():
    if not comprobar_porto(host_local, porto, nome):
        todos_dispoñibles = False



# Resulta das enquisas portuarias.

print("==============================")
if todos_dispoñibles:
    print("TODOS os servizos están listos para a conexión.")
else:
    print("\n⚠️⚠️⚠️ALGÚN servizo non está dispoñible.\nComprobe o estado de Docker.")
    '''Como estamos a empregar un Caderno de Jupyter e o cacho remata eiquí
    non é preciso empregar sys.exit() ou similar pra dete-la execución'''
    
````
